```python
#pip(3) install bs4 --> library to use beautifulsoup
#pip(3) install requests --> allow us to retrieve HTML content from website
#pip(3) install lxml --> web parser beauitfulSoup uses to retrieve content

from bs4 import BeautifulSoup
#requests --> HTTP requests you send to the website to get content
import requests

import pandas as pd
```

```python
website = "https://www.cnn.com/specials/space-science"
web = requests.get(website)
content = web.text


soup = BeautifulSoup(content, 'lxml')
print(soup.prettify())

#open full output in text editor to see h3 tag --> where titles are stored
```

```python
title = soup.find("title")
print(title.get_text())

#Modifying Tags
tag = soup.html
type(tag)
tag = "myTitle"
print(tag)
```

```python
headlines = soup.find_all("h3")
data = []
for line in headlines:
    data.append(line.getText())
    print("Headline : {}".format(line.getText()))
```

```python
df = pd.DataFrame({"News Headlines": data})
df
```

```python
df.to_csv('News_data.csv')
```

```python
#printing contents of tag

headline = soup.find("h3")
print(headline.contents)
print()
#printing parent of tag
print(headline.parent)
```

```python
#printing out headline
for news in soup.findAll(class_ = 'cn__column cn__column--2np1 cn__column--3np1 cn__column--4np3'):
    print(news.text.strip())
```

```python
#printing out content in children nodes
for content in soup.head.children:
    print(content)
```
